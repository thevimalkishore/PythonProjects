```python
def convert_length(value, from_unit, to_unit):
    units = {
        'm': 1,
        'km': 1000,
        'cm': 0.01,
        'mm': 0.001,
        'mi': 1609.34,
        'yd': 0.9144,
        'ft': 0.3048,
        'in': 0.0254
    }
    return value * units[from_unit] / units[to_unit]

def convert_weight(value, from_unit, to_unit):
    units = {
        'kg': 1,
        'g': 0.001,
        'mg': 0.000001,
        'lb': 0.453592,
        'oz': 0.0283495
    }
    return value * units[from_unit] / units[to_unit]

def convert_temperature(value, from_unit, to_unit):
    if from_unit == 'C':
        if to_unit == 'F':
            return (value * 9/5) + 32
        elif to_unit == 'K':
            return value + 273.15
    elif from_unit == 'F':
        if to_unit == 'C':
            return (value - 32) * 5/9
        elif to_unit == 'K':
            return (value - 32) * 5/9 + 273.15
    elif from_unit == 'K':
        if to_unit == 'C':
            return value - 273.15
        elif to_unit == 'F':
            return (value - 273.15) * 9/5 + 32
    return value

def main():
    print("Welcome to Python Unit Converter!")
    
    while True:
        print("\nSelect type of conversion:")
        print("1. Length")
        print("2. Weight")
        print("3. Temperature")
        print("4. Exit")
        
        choice = input("Enter choice (1/2/3/4): ")
        
        if choice == '4':
            print("Goodbye!")
            break
        
        value = float(input("Enter value: "))
        
        if choice == '1':
            from_unit = input("From unit (m, km, cm, mm, mi, yd, ft, in): ")
            to_unit = input("To unit (m, km, cm, mm, mi, yd, ft, in): ")
            result = convert_length(value, from_unit, to_unit)
        
        elif choice == '2':
            from_unit = input("From unit (kg, g, mg, lb, oz): ")
            to_unit = input("To unit (kg, g, mg, lb, oz): ")
            result = convert_weight(value, from_unit, to_unit)
        
        elif choice == '3':
            from_unit = input("From unit (C, F, K): ")
            to_unit = input("To unit (C, F, K): ")
            result = convert_temperature(value, from_unit, to_unit)
        
        else:
            print("Invalid choice, try again.")
            continue
        
        print(f"{value} {from_unit} = {result} {to_unit}")

if __name__ == "__main__":
    main()
```
