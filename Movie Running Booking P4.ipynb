```python
movies = {
    1: {"name": "Avengers", "price": 250},
    2: {"name": "KGF 2", "price": 200},
    3: {"name": "Spider-Man", "price": 300}
}


seats = ["Available"] * 10


def show_movies():
    print("\n=== MOVIES ===")
    for num, m in movies.items():
        print(f"{num}. {m['name']}  - ₹{m['price']}")


def show_seats():
    print("\n=== SEATS ===")
    for i in range(10):
        print(f"Seat {i+1}: {seats[i]}")


def book_ticket():
    show_movies()
    
    try:
        movie_choice = int(input("\nEnter movie number: "))
        if movie_choice not in movies:
            print("Invalid movie!")
            return
    except:
        print("Please enter a valid number!")
        return
    
    movie_name = movies[movie_choice]["name"]
    movie_price = movies[movie_choice]["price"]

    show_seats()
    try:
        seat_choice = int(input("\nEnter seat number (1-10): "))
        if seat_choice < 1 or seat_choice > 10:
            print("Invalid seat number!")
            return
    except:
        print("Please enter a valid number!")
        return

    if seats[seat_choice - 1] == "Booked":
        print("Seat already booked!")
        return

   
    seats[seat_choice - 1] = "Booked"

    
    print("\n======= RECEIPT =======")
    print(f"Movie      : {movie_name}")
    print(f"Seat       : {seat_choice}")
    print(f"Price      : ₹{movie_price}")
    print("========================\n")



while True:
    print("==== MOVIE BOOKING SYSTEM ====")
    print("1. Show Movies")
    print("2. Show Seats")
    print("3. Book Ticket")
    print("4. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        show_movies()

    elif choice == "2":
        show_seats()

    elif choice == "3":
        book_ticket()

    elif choice == "4":
        print("Thanks for using the system!")
        break

    else:
        print("Invalid choice! Try again.\n")
```
