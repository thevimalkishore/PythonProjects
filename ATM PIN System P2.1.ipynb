```python
PIN = "1987"      
balance = 9000      
attempts = 0

print("=== Welcome to Python ATM ===")

def print_receipt(transaction_type, amount, balance):
    from datetime import datetime
    print("\n---------- RECEIPT ----------")
    print(f"Transaction : {transaction_type}")
    print(f"Amount      : ₹{amount}")
    print(f"Balance     : ₹{balance}")
    print("Date & Time :", datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
    print("------------------------------\n")

while attempts < 3:
    user_pin = input("Enter your 4-digit PIN: ")

    if user_pin == PIN:
        print("\nPIN Verified Successfully!\n")

        while True:
            print("----- MENU -----")
            print("1. Check Balance")
            print("2. Deposit Money")
            print("3. Withdraw Money")
            print("4. Exit")

            choice = input("Select an option: ")

           
            if choice == "1":
                print(f"Your balance is: ₹{balance}\n")

           
            elif choice == "2":
                amount = float(input("Enter amount to deposit: "))
                balance += amount
                print(f"Deposited ₹{amount}. New balance: ₹{balance}")

                
                print_receipt("Deposit", amount, balance)

            
            elif choice == "3":
                amount = float(input("Enter amount to withdraw: "))

                if amount > balance:
                    print("Insufficient balance!\n")
                else:
                    balance -= amount
                    print(f"Withdrawn ₹{amount}. Remaining balance: ₹{balance}")

                   
                    print_receipt("Withdrawal", amount, balance)

           
            elif choice == "4":
                print("Thank you for using the ATM!")
                exit()

            else:
                print("Invalid option!\n")

    else:
        attempts += 1
        print("Incorrect PIN!")

        if attempts == 3:
            print("\nYour account has been locked due to 3 incorrect attempts.")
        else:
            print(f"Attempts remaining: {3 - attempts}")
```
